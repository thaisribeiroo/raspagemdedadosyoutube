```python
import requests
import csv
from langdetect import detect

API_KEY = "AIzaSyA_Dm4N-Uz8vsxa9C0sHCtDs993gNYSQqU"
PLAYLIST_ID = "PLc6EfuKFtZrLLHsGgC9yUTRNQxqB8mffk"

def get_playlist_videos(playlist_id):
    videos = []
    base_url = "https://www.googleapis.com/youtube/v3/playlistItems"
    params = {
        "part": "snippet",
        "playlistId": playlist_id,
        "maxResults": 50,
        "key": API_KEY
    }
    while True:
        res = requests.get(base_url, params=params)
        data = res.json()
        for item in data.get("items", []):
            video_id = item["snippet"]["resourceId"]["videoId"]
            videos.append(video_id)
        if "nextPageToken" in data:
            params["pageToken"] = data["nextPageToken"]
        else:
            break
    return videos

def get_videos_data(video_ids):
    url = "https://www.googleapis.com/youtube/v3/videos"
    all_video_data = []
    for i in range(0, len(video_ids), 50):
        batch_ids = video_ids[i:i+50]
        params = {
            "part": "snippet,statistics,contentDetails",
            "id": ",".join(batch_ids),
            "key": API_KEY
        }
        res = requests.get(url, params=params)
        data = res.json()
        all_video_data.extend(data.get("items", []))
    return all_video_data

def get_comments(video_id, max_comments=1000):
    comments = []
    url = "https://www.googleapis.com/youtube/v3/commentThreads"
    params = {
        "part": "snippet",
        "videoId": video_id,
        "maxResults": 100,
        "textFormat": "plainText",
        "key": API_KEY
    }
    while True:
        res = requests.get(url, params=params)
        data = res.json()
        for item in data.get("items", []):
            comment = item["snippet"]["topLevelComment"]["snippet"]
            comments.append({
                "author": comment["authorDisplayName"],
                "text": comment["textDisplay"],
                "likeCount": comment["likeCount"],
                "publishedAt": comment["publishedAt"]
            })
            if len(comments) >= max_comments:
                return comments
        if "nextPageToken" in data and len(comments) < max_comments:
            params["pageToken"] = data["nextPageToken"]
        else:
            break
    return comments

def is_portuguese(text):
    try:
        return detect(text) == "pt"
    except:
        return False

def main():
    video_ids = get_playlist_videos(PLAYLIST_ID)
    print(f"Total vídeos encontrados: {len(video_ids)}")

    videos_data = get_videos_data(video_ids)

    with open("youtube_dados.csv", mode="w", newline="", encoding="utf-8") as csv_file:
        fieldnames = ["Video ID", "Título", "Likes", "Views", "Qtd Comentários", "Comentário (PT)", "Autor", "Likes do Comentário", "Data"]
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()

        for video in videos_data:
            video_id = video["id"]
            snippet = video["snippet"]
            stats = video.get("statistics", {})

            title = snippet["title"]
            like_count = stats.get("likeCount", 0)
            view_count = stats.get("viewCount", 0)
            comment_count = stats.get("commentCount", 0)

            comments = get_comments(video_id, max_comments=1000)
            comments_pt = [c for c in comments if is_portuguese(c["text"])]

            for comment in comments_pt:
                writer.writerow({
                    "Video ID": video_id,
                    "Título": title,
                    "Likes": like_count,
                    "Views": view_count,
                    "Qtd Comentários": comment_count,
                    "Comentário (PT)": comment["text"],
                    "Autor": comment["author"],
                    "Likes do Comentário": comment["likeCount"],
                    "Data": comment["publishedAt"]
                })
    print("✅ Dados salvos em 'youtube_dados.csv'.")

if __name__ == "__main__":
    main()
```

```python
# prompt: Qual o total de views ao longo do tempo? Crie um gráfico de linhas com os números em evidência.

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Carregar os dados do CSV
df = pd.read_csv('youtube_dados.csv')

# Converter a coluna 'Views' para numérica (lidando com possíveis valores não numéricos)
df['Views'] = pd.to_numeric(df['Views'], errors='coerce').fillna(0)

# Converter a coluna 'Data' para datetime e ordenar
df['Data'] = pd.to_datetime(df['Data'])
df = df.sort_values(by='Data')

# Agrupar por data e somar as views para obter o total de views por dia
total_views_por_data = df.groupby(df['Data'].dt.date)['Views'].sum().reset_index()

# Criar o gráfico de linhas
plt.figure(figsize=(15, 7))
plt.plot(total_views_por_data['Data'], total_views_por_data['Views'], marker='o', linestyle='-')

# Adicionar os números de views em evidência
for i, row in total_views_por_data.iterrows():
    plt.text(row['Data'], row['Views'], f'{row["Views"]:.0f}', ha='center', va='bottom')

# Formatar o eixo Y como números inteiros e sem notação científica
plt.gca().yaxis.set_major_formatter(mtick.FormatStrFormatter('%.0f'))
plt.ticklabel_format(style='plain', axis='y')

# Configurar o título e rótulos dos eixos
plt.title('Total de Views ao Longo do Tempo', fontsize=16)
plt.xlabel('Data', fontsize=12)
plt.ylabel('Total de Views', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
# prompt: gere um grafo dos temas mais citados

!pip install networkx

import networkx as nx

def create_theme_graph(df, column='Comentário (PT)', min_frequency=5):
    """
    Gera um grafo onde os nós são os temas (palavras) mais citados
    e as arestas representam co-ocorrência desses temas nos comentários.

    Args:
        df (pd.DataFrame): DataFrame contendo os comentários.
                           Deve ter a coluna especificada por `column`.
        column (str): O nome da coluna que contém os comentários em português.
                      Default é 'Comentário (PT)'.
        min_frequency (int): Frequência mínima para uma palavra ser considerada um "tema".
    """
    if df.empty or column not in df.columns:
        print(f"DataFrame está vazio ou a coluna '{column}' não existe.")
        return

    # Combina todos os comentários em uma única string para análise de frequência geral
    text = ' '.join(df[column].dropna().astype(str))

    if not text.strip():
        print(f"A coluna '{column}' está vazia ou contém apenas espaços em branco.")
        return

    # Define stopwords em português e adiciona algumas comuns que podem não estar na lista padrão
    stopwords_pt = set(stopwords.words('portuguese'))
    stopwords_pt.update(['pra', 'vc', 'q', 'tbm', 'msm', 'tipo', 'né', 'vou', 'daí', 'então', 'muito', 'mais', 'vai', 'sim', 'tá', 'ter', 'fazer', 'dizer', 'vir', 'ver']) # Adicionei mais algumas comuns

    # Adiciona pontuação para remoção
    punctuation = set(string.punctuation)

    # Pré-processamento do texto para identificação de temas
    def preprocess_text_for_themes(text):
        text = text.lower()
        # Substituir pontuação por espaços para que as palavras adjacentes não fiquem coladas
        for p in punctuation:
            text = text.replace(p, ' ')
        tokens = word_tokenize(text, language='portuguese')
        # Remover stopwords, tokens não alfabéticos e tokens com 1 ou 2 caracteres
        filtered_tokens = [word for word in tokens if word not in stopwords_pt and word.isalpha() and len(word) > 2]
        return filtered_tokens

    # Processa todos os comentários para obter uma lista de tokens por comentário
    processed_comments_tokens = [preprocess_text_for_themes(str(comment)) for comment in df[column].dropna()]

    # Conta a frequência de cada token
    all_tokens = [token for sublist in processed_comments_tokens for token in sublist]
    token_counts = Counter(all_tokens)

    # Identifica os temas (palavras que aparecem com frequência mínima)
    themes = [token for token, count in token_counts.items() if count >= min_frequency]

    if not themes:
        print(f"Nenhum tema encontrado com frequência mínima de {min_frequency}. Considere diminuir o valor.")
        return

    # Cria o grafo
    G = nx.Graph()

    # Adiciona os temas como nós
    for theme in themes:
        G.add_node(theme, frequency=token_counts[theme])

    # Adiciona arestas com base na co-ocorrência nos comentários
    for tokens_in_comment in processed_comments_tokens:
        # Filtra os tokens do comentário para incluir apenas os que são temas
        comment_themes = [token for token in tokens_in_comment if token in themes]
        # Cria arestas entre todos os pares de temas que aparecem no mesmo comentário
        for i in range(len(comment_themes)):
            for j in range(i + 1, len(comment_themes)):
                node1 = comment_themes[i]
                node2 = comment_themes[j]
                # Adiciona ou incrementa o peso da aresta (contagem de co-ocorrência)
                if G.has_edge(node1, node2):
                    G[node1][node2]['weight'] += 1
                else:
                    G.add_edge(node1, node2, weight=1)

    # Remove nós isolados (temas que não co-ocorrem com outros temas acima do limiar)
    isolated_nodes = list(nx.isolates(G))
    G.remove_nodes_from(isolated_nodes)

    if G.number_of_nodes() == 0:
        print("Nenhum par de temas co-ocorreu. O grafo resultante está vazio.")
        return

    print(f"Grafo criado com {G.number_of_nodes()} nós (temas) e {G.number_of_edges()} arestas.")


    # Visualização do grafo
    plt.figure(figsize=(15, 15))

    # Usa um layout que se adapta a grafos
    pos = nx.spring_layout(G, k=0.8, iterations=50) # k controla a distância entre os nós

    # Desenha os nós
    # Tamanho do nó pode ser baseado na frequência do tema
    node_size = [G.nodes[node]['frequency'] * 50 for node in G.nodes()]
    nx.draw_networkx_nodes(G, pos, node_size=node_size, node_color='skyblue', alpha=0.9)

    # Desenha as arestas
    # Espessura da aresta pode ser baseada na co-ocorrência
    edge_width = [edge[2]['weight'] * 0.5 for edge in G.edges(data=True)]
    nx.draw_networkx_edges(G, pos, width=edge_width, alpha=0.5, edge_color='gray')

    # Desenha os rótulos dos nós
    nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

    plt.title('Grafo de Co-ocorrência de Temas nos Comentários')
    plt.axis('off') # Esconde os eixos
    plt.show()

# Exemplo de uso com o DataFrame existente 'df'
if not df.empty:
    print("\nGerando Grafo de Temas Mais Citados:")
    # Ajuste min_frequency conforme a quantidade e a natureza dos seus dados
    # Um valor menor inclui mais palavras no grafo
    create_theme_graph(df, column='Comentário (PT)', min_frequency=10)
else:
    print("DataFrame está vazio. Não foi possível gerar o grafo de temas.")
```

```python
# prompt: preciso de um grafo que mostre os sentimentos dos comentários e como estão interligados

import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string

# Certifique-se de que os recursos do NLTK estão baixados
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)

def preprocess_text_for_graph(text):
    """Pré-processa o texto para extrair tokens relevantes para o grafo."""
    text = text.lower()
    # Substituir pontuação por espaços
    for p in string.punctuation:
        text = text.replace(p, ' ')
    tokens = word_tokenize(text, language='portuguese')

    # Define stopwords em português
    stopwords_pt = set(stopwords.words('portuguese'))
    stopwords_pt.update([
        'pra', 'vc', 'q', 'tbm', 'msm', 'tipo', 'né', 'vou', 'daí', 'então',
        'muito', 'mais', 'vai', 'sim', 'tá', 'ter', 'fazer', 'dizer', 'vir',
        'ver', 'gente', 'coisa', 'aí', 'lá', 'aqui', 'agora', 'sempre',
        'cada', 'todo', 'toda', 'tudos', 'todas', 'ficou', 'ficar', 'fala',
        'falar', 'acho', 'achar', 'quase', 'apenas', 'bom', 'boa', 'ruim',
        'ótimo', 'ótima', 'obrigado', 'obrigada', 'valeu', 'pena', 'super',
        'melhor', 'pior', 'assim', 'ser', 'estar', 'ir', 'poder', 'dar',
        'ter', 'haver', 'vir', 'ver', 'saber', 'fazer', 'dizer', 'querer',
        'dever', 'usar', 'parece', 'parecer' # Adicione mais stopwords comuns e genéricas
    ])

    # Remover stopwords, tokens não alfabéticos, tokens com 1 ou 2 caracteres
    # e tokens que são apenas dígitos
    filtered_tokens = [
        word for word in tokens
        if word not in stopwords_pt and word.isalpha() and len(word) > 2
    ]
    return filtered_tokens

def create_sentiment_cooccurrence_graph(df, sentiment_label, min_frequency=5, edge_weight_threshold=2):
    """
    Cria um grafo de co-ocorrência para comentários com um sentimento específico.

    Args:
        df (pd.DataFrame): DataFrame contendo os comentários e a coluna 'Sentiment'.
        sentiment_label (str): O rótulo de sentimento para filtrar ('Positivo', 'Negativo', 'Neutro').
        min_frequency (int): Frequência mínima para uma palavra ser considerada um nó no grafo.
        edge_weight_threshold (int): Peso mínimo (co-ocorrência) para uma aresta ser incluída.

    Returns:
        networkx.Graph: O grafo de co-ocorrência para o sentimento especificado.
    """
    if df.empty or 'Comentário (PT)' not in df.columns or 'Sentiment' not in df.columns:
        print("DataFrame está vazio ou colunas necessárias não existem.")
        return nx.Graph()

    # Filtra os comentários pelo sentimento desejado
    comments_by_sentiment = df[df['Sentiment'] == sentiment_label]['Comentário (PT)'].dropna().tolist()

    if not comments_by_sentiment:
        print(f"Não há comentários com o sentimento '{sentiment_label}'.")
        return nx.Graph()

    # Processa os comentários para obter tokens
    processed_comments_tokens = [preprocess_text_for_graph(str(comment)) for comment in comments_by_sentiment]

    # Conta a frequência de cada token em todos os comentários filtrados
    all_tokens = [token for sublist in processed_comments_tokens for token in sublist]
    token_counts = Counter(all_tokens)

    # Identifica as palavras que serão nós do grafo (com frequência mínima)
    graph_nodes = [token for token, count in token_counts.items() if count >= min_frequency]

    if not graph_nodes:
        print(f"Nenhum termo encontrado com frequência mínima de {min_frequency} para o sentimento '{sentiment_label}'.")
        return nx.Graph()

    # Cria o grafo
    G = nx.Graph()

    # Adiciona os nós
    for node in graph_nodes:
        G.add_node(node, frequency=token_counts[node])

    # Adiciona arestas com base na co-ocorrência dentro de cada comentário
    for tokens_in_comment in processed_comments_tokens:
        # Filtra os tokens do comentário para incluir apenas os que são nós do grafo
        comment_nodes = [token for token in tokens_in_comment if token in graph_nodes]
        # Cria arestas entre todos os pares de nós que aparecem no mesmo comentário
        for i in range(len(comment_nodes)):
            for j in range(i + 1, len(comment_nodes)):
                node1 = comment_nodes[i]
                node2 = comment_nodes[j]
                # Adiciona ou incrementa o peso da aresta (contagem de co-ocorrência)
                if G.has_edge(node1, node2):
                    G[node1][node2]['weight'] += 1
                else:
                    G.add_edge(node1, node2, weight=1)

    # Remove arestas com peso abaixo do limiar
    edges_to_remove = [(u, v) for u, v, data in G.edges(data=True) if data['weight'] < edge_weight_threshold]
    G.remove_edges_from(edges_to_remove)

    # Remove nós isolados após a remoção das arestas
    isolated_nodes = list(nx.isolates(G))
    G.remove_nodes_from(isolated_nodes)


    return G

# --- Geração e Visualização dos Grafos por Sentimento ---

if not df.empty and 'Sentiment' in df.columns:
    sentiments_to_analyze = ['Positivo', 'Negativo', 'Neutro']

    for sentiment in sentiments_to_analyze:
        print(f"\nGerando Grafo de Co-ocorrência para Sentimento: {sentiment}")
        # Ajuste min_frequency e edge_weight_threshold conforme seus dados e o nível de detalhe desejado
        G = create_sentiment_cooccurrence_graph(df, sentiment, min_frequency=10, edge_weight_threshold=3) # Ajuste esses valores!

        if G and G.number_of_nodes() > 0:
            plt.figure(figsize=(12, 12))

            # Usa um layout que se adapta a grafos
            pos = nx.spring_layout(G, k=0.9, iterations=50, seed=42) # k controla a distância, seed para reprodutibilidade

            # Desenha os nós
            # Tamanho do nó pode ser baseado na frequência do tema
            node_size = [G.nodes[node].get('frequency', 1) * 80 for node in G.nodes()] # Use .get com valor padrão
            # Cor dos nós pode ser baseada no sentimento
            node_color = {
                'Positivo': 'lightgreen',
                'Negativo': 'salmon',
                'Neutro': 'lightgray'
            }.get(sentiment, 'skyblue') # Cor padrão se o sentimento não estiver mapeado

            nx.draw_networkx_nodes(G, pos, node_size=node_size, node_color=node_color, alpha=0.9)

            # Desenha as arestas
            # Espessura da aresta pode ser baseada na co-ocorrência
            edge_width = [edge[2].get('weight', 1) * 0.8 for edge in G.edges(data=True)] # Use .get com valor padrão
            nx.draw_networkx_edges(G, pos, width=edge_width, alpha=0.6, edge_color='darkgray')

            # Desenha os rótulos dos nós
            nx.draw_networkx_labels(G, pos, font_size=9, font_weight='bold')

            plt.title(f'Grafo de Co-ocorrência de Termos - Sentimento: {sentiment}')
            plt.axis('off') # Esconde os eixos
            plt.show()
        elif G is not None: # Se G foi criado mas está vazio
             print(f"O grafo para o sentimento '{sentiment}' está vazio após a remoção de nós isolados/arestas de baixo peso.")
        # Se G é None, a mensagem de "Não há comentários" ou "Nenhum termo encontrado" já foi exibida.

else:
    print("DataFrame está vazio ou a coluna 'Sentiment' não existe. Execute a análise de sentimento primeiro.")

```
